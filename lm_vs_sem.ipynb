# Analysis of the difference in results from LM, SEM and litograph

## Intro
The values of the wire+gap width (further Width) comming from the 
* litograph software (LIT)
* light microscope (LM)
* SEM

contradict each other. 
Namely, the width set in litograph is 2.4 um is confirmed by the light microscope, but the Width measured in the SEM is significantly lower than this value.
The exact data follows.
